```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Load the dataset
url = 'https://raw.githubusercontent.com/sumankumarsubudhi/Spam-Classifier/master/SMSSpamCollection'
df = pd.read_csv(url, sep='\t', header=None, names=['label', 'message'])
```

# **Data Cleaning and Pre-processing**

```python
# Data Cleaning and Pre-processing
nltk.download('stopwords')
```

```python
ps = PorterStemmer()
```

```python
df['message'][0]
```

**Step: 1 -->  removing any characters that are not alphabets**

```python
# Cleaning using Regular expression
def clean_text(text):
    text = re.sub('[^a-zA-z]',' ',text)
    return text

df['message'] = df['message'].apply(clean_text)
```

```python
df['message'][0]
```

**Lower case all the text**

```python
df['message'] = df['message'].str.lower()
```

```python
df['message'][0].split()
```

```python
w = df['message'][0].split()
w
```

```python
' '.join(w)
```

**Removing all the stopwords**

```python
corpus = []

for i in range(len(df)):
    words = df['message'][i].split()
    words = [ps.stem(word) for word in words if word not in stopwords.words('english')]
    words = ' '.join(words)
    corpus.append(words)
```

```python
corpus
```

# **Creating the Bag of Words**

**Apply BOW, TF-IDF, Machine Learning algorithm**

        1. Cleaning and Preprocessing 
        2. Train Test Split
        3. BOW and TF-IDF (Sentence---->vectors)
        4.Trained Our Model

We have to first train test split and then apply BOW and TF-IDF to **Prevent the Data Leakage**. Like your test data have no clue about your train data

```python
X = corpus
```

```python
# Output feature
y = pd.get_dummies(df['label'],drop_first=True,dtype='int')
```

```python
y = y.values
```

```python
## Train test split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
# Creating a bag of words model
# For Binary bag of words
cv = CountVectorizer(max_features=2500,ngram_range=(1,2))
```

```python
len(X_train),len(y_train)
```

```python
# Independent Features
X_train = cv.fit_transform(X_train).toarray()
X_test = cv.transform(X_test).toarray()
```

```python
X_train
```

```python
cv.vocabulary_
```

```python
# Output features
df.head()
```

```python
spam_delect_model = MultinomialNB()
```

```python
spam_delect_model.fit(X_train,y_train)
```

```python
y_pred = spam_delect_model.predict(X_test)
```

```python
mse = accuracy_score(y_test,y_pred)
mse
```

```python
print(classification_report(y_test,y_pred))
```

# **Create the TF-IDF Model**

```python
X = corpus
```

```python
# Output feature
y = pd.get_dummies(df['label'],drop_first=True,dtype='int')
```

```python
y = y.values
```

```python
## Train test split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
tfidf = TfidfVectorizer(max_features=2500,ngram_range=(1,2))
```

```python
len(X_train),len(y_train)
```

```python
# Independent Features
X_train = tfidf.fit_transform(X_train).toarray()
X_test = tfidf.transform(X_test).toarray()
```

```python
X_train
```

```python
tfidf.vocabulary_
```

```python
spam_tfidf_model = MultinomialNB().fit(X_train,y_train)
```

```python
y_pred = spam_tfidf_model.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
